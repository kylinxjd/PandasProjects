```python
group_obj = d.groupby(['cumid', 'type'])
```

```python
r = group_obj[['time']].min()
r
```

```python
r_trans = pd.pivot_table(r, index='cumid', columns='type', values='time')
```

```python
r_trans
```

```python
r_trans = r_trans.fillna({
    'Normal':0,
    'Presented':0,
    'Special_offer':0,
    'returned_goods':0
})
r_trans
```

```python
r_trans['r_max'] = r_trans.apply(lambda x : x.max(), axis=1)
```

```python
r_trans
```
