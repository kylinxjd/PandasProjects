```python
d1 = pd.DataFrame({
    'animal': ['dog', 'dog', 'cat','cat'],
    'weight': [12, 15, 3, 4],
    'age':[12,6,16,2],
    'type':['哈士奇', '柴犬', '蓝猫', '蓝猫']
})
d1
```

```python
d1.groupby(by='animal').sum()
```

```python
d1.groupby(by='animal').mean()
```

```python
# 方差
d1.groupby(by='animal').var()
```

```python
# 标准差
d1.groupby(by='animal').std()
```

```python
d1.groupby(by='animal').max()
```

```python
d1.groupby(by='animal').min()
```

```python
# 中位数
d1.groupby(by='animal').median()
```

```python
d1.groupby(by='animal').size()
```

```python
d1.groupby(by='animal').count()
```

```python
obj = d1.groupby(by='animal')
for name, data in obj:
    print(name)
    print(data)
```

```python
dic_obj = dict(list(obj))
dic_obj
```

```python
d1.groupby(by='animal').describe()
```

```python
d1.groupby(by='animal').agg('max')
```

```python
d1.groupby(by='animal').agg([('最大值', 'max'), 
                             ('最小值', 'min'), 
                             ('平均值', 'mean')])
```

```python
def func(x):
#     print("--------------------")
    return np.max(x)-np.min(x)
d1.groupby(by='animal').agg([('最大差值', func)])
```
