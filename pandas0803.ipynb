```python
import pandas as pd
import numpy as np
data = pd.read_excel('豆瓣电影数据.xlsx', index_col=0, encoding='str')
df = data.head(100)
```

```python

df
```

## 透视表

```python
d = df
d
```

```python
# 以产地为索引，计算各个数字字段的均值
d.loc[:,'时长'] = d.loc[:,'时长'].astype(np.int32)
pd.pivot_table(d, index=['产地'])
```

```python
# 设置多索引
pd.pivot_table(d, index=['产地', '年代'])
```

```python
# 指定要统计的列数据
pd.pivot_table(d, index=['产地', '年代'], values=['评分'])
```

```python
# 指定函数，统计和
pd.pivot_table(d, index=['产地', '年代'], values=['投票人数'], aggfunc=np.sum)
```

```python
# 同时计算多个统计量
pd.pivot_table(d, index=['产地'], values=['投票人数','评分'], aggfunc=[np.sum,np.mean])
```

```python
dic = {
    '名字':'复仇者联盟3',
    '投票人数':4565142,
    '类型':'剧情/科幻',
    '产地':'美国',
    '上映时间':'1994-09-10 10:00:00',
    '时长':np.nan,
    '年代':np.nan,
    '评分':np.nan,
    '首映地点':'美国'
}
s = pd.Series(dic)
s.name = 1024
d = d.append(s)
```

```python
d
```

```python
# 空值用0替换
pd.pivot_table(d, index=['产地'], values=['投票人数','评分'], aggfunc=[np.sum,np.mean], fill_value=0)
```

```python
# 在统计结果下方显示总和
pd.pivot_table(d, index=['产地'], values=['投票人数','评分'], aggfunc=[np.sum,np.mean], fill_value=0, margins=True)
```

```python
# 对不同的值使用不同的函数
pd.pivot_table(d, index=['产地'], values=['投票人数','评分'], 
               aggfunc={
                   '投票人数':np.sum,
                   '评分':np.mean
               }, 
               fill_value=0)
```

```python
# 透视表过滤
table = pd.pivot_table(d, index=['年代'], values=['投票人数','评分'], 
               aggfunc={
                   '投票人数':np.sum,
                   '评分':np.mean
               }, 
               fill_value=0)
```
