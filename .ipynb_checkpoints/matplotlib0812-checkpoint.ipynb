```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
snd = pd.read_csv('sndHsPr.csv')
da = snd
```

```python
da['total_price'] = da['price']*da['AREA']
# da['total_price'] = da[['price','AREA']].apply(lambda x: x[0]*x[1], axis=1)
```

```python
da
```

```python
da.groupby('dist').count()
```

```python
def ch(x):
    if x =='chaoyang':
        return '朝阳区'
    elif x == 'dongcheng':
        return '东城区'
    elif x == 'fengtai':
        return '丰台区'
    elif x == 'haidian':
        return '海淀区'
    elif x == 'shijingshan':
        return '石景山区'
    elif x == 'xicheng':
        return '西城区'
label = {
    'chaoyang':'朝阳',
    'dongcheng':'东城',
    'fengtai':'丰台',
    'haidian':'海淀',
    'shijingshan':'石景山'
}
da['name'] = da['dist'].apply(ch)
```

```python
da
```

```python
avg_price = da.groupby('name').agg('mean')['price']
avg_price
```

```python
# 修改字体，支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']
# 当字体是中文的时候显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(avg_price.index, avg_price.values)
```

```python
plt.figure(figsize=(18, 12))
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('地区', fontsize=24)
plt.ylabel('价格', fontsize=28)
plt.bar(avg_price.index, avg_price.values, )
for a, b in zip(avg_price.index, avg_price.values):
    plt.text(a, b+550, round(b, 3), ha='center', va='bottom', fontsize=20)
```
