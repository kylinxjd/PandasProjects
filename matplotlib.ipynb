```python
import matplotlib.pyplot as plt
import warnings
import matplotlib

warnings.filterwarnings('ignore')
# # 设置字体显示中文
# plt.rcParams['font.sans-serif'] = ['SimHei']
# # 设置显示负号等字符
# plt.rcParams['axes.unicode_minus'] = False

# 方法二
# font = {
#     'family':'MicroSoft YaHei',
#     'weight': 'bold',
#     'size': 12
# }
# matplotlib.rc('font', **font)

plt.figure(num=1, figsize=(10, 8), dpi=80)
# dpi=80 设置图片清晰度，每英寸80个像素点

# 设置x轴刻度
plt.xticks([i for i in range(1, 11)])

plt.plot([i for i in range(1, 11)], [4, 5, 8, 6, 3, 5,6,4,2,8])
plt.xlabel('x轴', fontsize=16)
plt.ylabel('y轴', fontsize=16)
# plt.show()
```

```python
import random
plt.figure(num=2, figsize=(10, 8), dpi=80)
x = range(0, 120)
y = [random.randint(20, 35) for i in range(120)]

# 自定义 X 轴
_x = list(x)[::10]
plt.xticks(_x, ['hello-{}'.format(i) for i in _x], rotation=45)

plt.title('按时到货计划时间的反馈', fontsize=18)

plt.plot(x, y) 
```

```python
plt.figure(num=3, figsize=(18, 8), dpi=80)

x = range(11, 31)
y = [random.randint(0, 6) for i in range(20)]
y2 = [random.randint(0, 6) for i in range(20)]

plt.plot(x, y, 'c', linewidth=5, label='男')
plt.plot(x, y2, 'r', label='女')

x_label = ['{}岁'.format(i) for i in x]

plt.xticks(x, x_label)

# 绘制网格
plt.grid(alpha=0.4)

# 图示
plt.legend(loc=0)

plt.show()
```

## 多条形图

```python
x1 = range(1, 6)

y1 = [random.randint(1, 16) for i in range(5)]
y2 = [random.randint(1, 16) for i in range(5)]
y3 = [random.randint(1, 16) for i in range(5)]

plt.figure(num=7, figsize=(18, 12), dpi=80)
# width=0.5  柱子宽度
bar_width = 0.3

x2 = [x-bar_width for x in x1]

x3 = [x+bar_width for x in x1]

plt.bar(x1, y1, color='red', width=bar_width, label='sss-1')
plt.bar(x2, y2, color='cyan', width=bar_width, label='sss-2')
plt.bar(x3, y3, color='pink', width=bar_width, label='sss-3')

plt.xticks(list(range(6)))
plt.xlabel('沙发沙发', fontsize=18 )
plt.ylabel('否合法就', fontsize=18)

plt.legend()

plt.show()
```
